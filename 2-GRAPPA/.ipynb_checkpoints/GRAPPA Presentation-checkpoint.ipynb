```julia
using Distributions
using Random
using PyPlot
using FFTW
using Images
using Base
using LinearAlgebra
using DSP
```

```julia
X = ones(500,500);
Y = ones(500,500);
for i=1:500
    X[:,i] = X[:,i].*i;
    Y[i,:] = Y[i,:].*i;
end
```

```julia
gausX = pdf.(Normal(250,180),X)
gausY = pdf.(Normal(450,200),Y);
gaus = gausX.*gausY;
imshow(gaus, cmap="hot");
```

```julia
fourier = FFTW.fft(gaus) ;
fshift = FFTW.fftshift(fourier)
magnitude =  log.(abs.(fshift));
```

```julia
imshow(magnitude, cmap="gray")
```

```julia
N = 500
shepp = shepp_logan(N, highContrast=true);
```

```julia
imshow(shepp.*gaus, cmap="gray")
```

```julia
fourier = FFTW.fft(shepp.*gaus) ;
fshift = FFTW.fftshift(fourier)
magnitude =  log.(abs.(fshift));
```

```julia
imshow(magnitude, cmap="gray");
```

```julia
gausX2 = pdf.(Normal(250,180),X)
gausY2 = pdf.(Normal(50,200),Y);
gaus2 = gausX2.*gausY2;
imshow(gaus2, cmap="hot");
```

```julia
fourier_shepp = FFTW.fft(shepp) ;
fshift_shepp = FFTW.fftshift(fourier_shepp)
magnitude =  log.(abs.(fshift_shepp));
```

```julia
k_coil1 = FFTW.fftshift(FFTW.fft(shepp));
k_coil2 = FFTW.fftshift(FFTW.fft(shepp.*gaus2));
```

```julia
X_train = zeros(ComplexF64, (6,71*498));
y_train = zeros(ComplexF64, (71*498,1));
idx = 1;
for z=210:280
    for p=2:499
        a,b,c = fshift_shepp[z-1, [p-1,p,p+1]]
        d,e,f = fshift_shepp[z+1, [p-1,p,p+1]]
        X_train[:,idx] = [a,b,c,d,e,f]
        y_train[idx] = fshift_shepp[z,p]
        idx +=1;
    end
end
```

```julia
@show size(X_train)
@show size(y_train);
```

```julia
W = pinv(X_train)' *y_train
```

```julia
sum(isless.(abs.(k_coil2) , 1e-5 ))
```

```julia
maximum(abs.(k_coil1))
```
