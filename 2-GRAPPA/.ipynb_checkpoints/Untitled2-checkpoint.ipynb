```julia
# Probably need to rerun a few time to see a nice reconstruction, 
# as only a few coeffiencts are used and not likely not the optimal ones either

# We pretend we do not know above function and its FT but 
# to simulate nfft we pretend to have aquired some fourier 
# coeffiencts of above function in the k-space 
v = rand(Bool,2 * f_s) # Create a random vector of Booleans to choose fourier coeffiecents and respective frequencies
fHat = X[v]            # Extract Coeffiecents
xx = freqs[v]          # Extract Frequencies
figure(figsize=(4,4))
plot(xx,abs.(fHat),"+")
xlabel("freq.")
ylabel("Amp.")
title("aquired fourier coefficents")
```

```julia
# Perform NFFT
p = NFFTPlan(xx, 2*f_s);              # create plan. m and sigma are optional parameters
f = nfft_adjoint(p, fHat);            # calculate adjoint NFFT aka reconstruct
g = nfft(p, f);                      # compute NFFT 
```

```julia
figure(figsize=(12,4))
subplot(1,2,1)
plot(t,real.(f))
xlabel("t")
ylabel("y")
title("recon. function")
subplot(1,2,2)
plot(xx,abs.(g),"+")
xlabel("freq.")
ylabel("Amp.")
title("nfft of recon. function")
```
