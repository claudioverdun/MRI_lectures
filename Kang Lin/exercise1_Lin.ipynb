#### Task 1

The radio spectrum is the part of the electromagnetic spectrum with frequencies from 30 Hz to 300 GHz. MRI scanners detect hydrogen nuclei in the water molecules in the human body. A carefully chosen radiofrequency pulse is applied, which causes the hydrogen nuclei to flip over. As the hydrogen nuclei move back to their eqilibrium state, they emit radio signals at the same frequency as the pulse they just absorbed. This phenomenon is called nuclear magnetic resonance. The radiofrequency coils in MRI machines receive those signals, which are then transformed into a visible images.

Any square matrix $R \in \mathbb{R}^n$ is a rotation matrix, if and only if $R$ is a orthogonal matrix and has determinant $|R| = 1$.
Let  $\omega = \gamma B_0$, and

$$
R = \left(\begin{array}{ccc} 
\cos(\omega t) & \sin(\omega t) & 0 \\
-\sin(\omega t) & \cos(\omega t) & 0 \\
0 & 0 & 1 \\
\end{array}\right),
$$

then matrix $R$ is a rotation matrix as $R^\top R = R R^\top = I$ and $|R| = 1$. Furthermore, given any vector 
$v = \left(\begin{array}{c} 
x \\
y \\
z \\
\end{array}\right)
$ we can compute

$$
Rv = \left(\begin{array}{c} 
R_{xy} \left(\begin{array}{c} x \\ y \end{array}\right)\\
z \\
\end{array}\right),
$$

with $R_{xy} = \left(\begin{array}{cc} 
\cos(\omega t) & \sin(\omega t) \\
-\sin(\omega t) & \cos(\omega t)  \\
\end{array}\right) $. One can easily check and confirm that also $R_{xy}$ is also a rotation matrix. Since applying $Rv$ doesn't affect the z coordinate and $R_{xy}$ only affects the x and y coordinates, we conclude that $R$ is a rotation around the z-axis.

T2 is the transverse or spin-spin relaxation time, while T1 is the longitudinal spin-lattice relaxation time. T2 and T1 are both time constants. The transverse magnetization vector drops to $\frac{1}{e} \approx 37\%$ of its original magnitude after one time constant T2. The z component of the nuclear spin magnetization recovers to $1 - \frac{1}{e} \approx 63\%$ of its equilibrium value after one time constant T1. So, T2 is associated with the component in $M$ that is perpendicular to $B_0$, which emerged from flipping the hydrogen nucleus. Hence, T2 is shorter than T1 because analogous to flushing a toilet, the flush duration is much shorter then the refill duration.

```julia
using Images
using FFTW
using Wavelets
using PyPlot
```

```julia
# Function that does all the transformation asked in the exercise
function trans(img)
    F = fft(img);
    F = fftshift(fft(img));
    
    W = dwt(img, wavelet(WT.db2));
    
    D = dct(img);
    
    return F, W, D
    
end;
```

In each image we see that both transformations have lots of small coefficients close to zero and much fewer coeffiencts that are significantly larger, though, the DWT in general yields much smaller values (even exact 0) than DCT. Furthermore, we see that for both transformations the amount-change from small to large coefficients is similarly steep. Also in both cases there seem to be a rather distinct level at  $\approx$ 0.1, which tells me that if I dissmiss all coefficients smaller than 0.1, I can reconstruct a nice image. However, since DWT yields many more significantly smaller coefficients than DCT, i.e. dismissing those have smaller impact, I suppose that DWT would perform better as sparse representation for both images.

So lets check in the following by dismissing all coefficients smaller than 0.1

```julia
function task10_recon(img,a,b)    
    figure(figsize=(10,10))
    img = Gray.(img)
    img = convert(Array{Float64}, img);
    _ , W, D = trans(img);
    W_abs, D_abs = abs.(W), abs.(D);
    subplot(4,3,1);
    imshow(img,cmap="gray")
    axis("off")
    title("Original")

    # Reconstruct with sparse DWT
    W[abs.(W).<a] .= 0
    im = idwt(W, wavelet(WT.db2))


    subplot(4,3,2);
    imshow(abs.(im),cmap="gray") 
    axis("off")
    title("Recon. iDWT")

    # Reconstruct with sparse DCT
    D[abs.(D).<b] .= 0
    im2 = idct(D)

    subplot(4,3,3);
    imshow(abs.(im2),cmap="gray") 
    axis("off")
    title("Recon. iDCT")

    # Ratio of coeficients used
    println("Amount of coeficients used for iDWT in % : ", round(sum(abs.(W).>=a)/(size(img)[1]*size(img)[2])*100, digits = 3))
    println("PSNR of iDWT: ", round(10*log10(maximum(img)^2/sum((im-img).^2)*(size(img)[1]*size(img)[2])),digits = 3))
    println("Amount of coeficients used for iDCT in % : ", round(sum(abs.(D).>=b)/(size(img)[1]*size(img)[2])*100, digits = 3))
    println("PSNR of iDCT: ", round(10*log10(maximum(img)^2/sum((im2-img).^2)*(size(img)[1]*size(img)[2])),digits = 3))
end;    
```

```julia
img = load("brain.jpg")
task10_recon(img,0.1,0.1) 
```

```julia
img = load("angiogram.jpg")
task10_recon(img,0.1,0.1) 
```

Here we see that DWT actually performs better as sparse representation in terms of PSNR, as for both images the inverse DWT is performed with fewer coefficients than with inverse DCT, yet yielding a slightly better PSNR. However, by subjectively looking at those images, it does seem that the sparse inverse DCT look a bit sharper even with less PSNR but at the cost of more coefficients.
